```python
import pandas as pd

# Cargar el archivo JSON
df = pd.read_json("../../data/raw/baches_2021.json")

# Mostrar las primeras filas
df.head(5)
```

```python
df.info()
```

```python
#droppeando description ya que es una columna repetida
df = df.drop(columns=['description'])
```

```python
#Revisando valores unicos de Material
df['material'].unique()
```

```python
# Droppeando material ya que no aporta informacion relevante
df = df.drop(columns=['material'])
```

```python
df.head()
```

```python
#Mappeando meses del año a numeros
month_map = { 'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4, 'Mayo': 5, 'Junio': 6,
              'Julio': 7, 'Agosto': 8, 'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12 }
```

```python
#viendo el tipo de dato que es fecha_reporte
df['fecha_reporte'].dtype
#convirtiendo a string
df['fecha_reporte'] = df['fecha_reporte'].astype(str)
#usando regex para eliminar cualquier caracrter que no sea una letra o u numero
df['fecha_reporte'] = df['fecha_reporte'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
```

```python
df['fecha_reporte'] = df['fecha_reporte'].astype(str)
```

```python
#Usando regex para cambiar el nombre del mes por su numero correspondiente
for month, num in month_map.items():
    df['fecha_reporte'] = df['fecha_reporte'].str.replace(month, str(num), regex=False)
```

```python
df.head(5)
```

```python
#separando fecha_reporte en dia, mes y año
df[['month', 'day', 'year']] = df['fecha_reporte'].str.split(' ', expand=True)
```

```python
#Eliminando columna fecha_reporte
df = df.drop(columns=['fecha_reporte'])
```

```python
#Recreando fecha_reporte en formato datetime
df['fecha_reporte'] = pd.to_datetime(df[['day', 'month', 'year']])
```

```python
df.head()
```

```python
#Cambiando el nombre del mes en fecha_atencion a su numero correspondeinte
df['fecha_atencion'] = df['fecha_atencion'].astype(str)
for month, num in month_map.items():
    df['fecha_atencion'] = df['fecha_atencion'].str.replace(month, str(num), regex=False)
```
